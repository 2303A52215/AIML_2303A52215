## Part 2 – Visualization of Data using matplotlib, pyplots Packages
lets suppose we collected age in "years", height in"cm" and weight in "kg" information few patients visiting a clinic


```python
import matplotlib.pyplot as plt
import numpy as np
name=["goku","gohan","naruto","luffy","zoro","ash"]
age=[45,25,19,19,20,10]
height=[160,150,155,170,180,130]
weight=[70,65,50,55,60,38]
```

### **Scatter graph**

```python
plt.scatter(weight,height)
plt.ylabel("height")
plt.xlabel("weight")
plt.title("height vs weight")
plt.show()
```

### ***bar graph***

```python
plt.figure(figsize=(12,4))
plt.title("Patient's weight in kgs",fontsize=16,
          fontstyle='italic')
plt.bar(x=name,height=weight, width=0.5,
        color='blue',edgecolor='black',alpha=1)
plt.xlabel("Patients",fontsize=15)
plt.xticks(fontsize=13,rotation=35,color="red")
plt.yticks(fontsize=13)
plt.ylabel("Weight (in kgs)",fontsize=15)
plt.show()
```

### **histogram**

```python
plt.figure(figsize=(6,6))
plt.hist(weight,color='b',edgecolor='red', alpha=1,bins=5)
plt.title("Histogram of patient weight",fontsize=18)
plt.xlabel("Weight in kgs",fontsize=12)
plt.xticks(fontsize=12)
plt.yticks(fontsize=12)
plt.show()
```
